```python
df.fillna(df['sepalwidth'].mean())
```

```python
df.fillna(df['sepalwidth'].mode())
```

```python
q1 = df['sepalwidth'].quantile(0.25)
q3 = df['sepalwidth'].quantile(0.75)

iqr = q3 - q1

low = q1 - 1.5*iqr
high = q3 + 1.5*iqr

df = df.loc[(df['sepalwidth'] > low) & (df['sepalwidth'] < high)]
```

```python
sns.boxplot(data = df)
```

```python
df['petallength'] = df['petallength'] / df['petallength'].max()
```
