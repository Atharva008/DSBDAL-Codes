```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris.csv')
df1 = df.copy()
df2 = df.copy()
```

```python
df1
```

```python
sns.boxplot(data = df1)
```

```python
df1['sepalwidth'] = (df1['sepalwidth'] - df1['sepalwidth'].min()) / (df1['sepalwidth'].max() - df1['sepalwidth'].min())
```

```python
df1
```

```python
sns.boxplot(data = df1)
```

```python
q1 = df['sepalwidth'].quantile(0.25)
q3 = df['sepalwidth'].quantile(0.75)

iqr = q3 - q1

low = q1 - 1.5*iqr
high = q3 + 1.5*iqr

df = df.loc[(df['sepalwidth'] > low) & (df['sepalwidth'] < high)]
```

```python
sns.boxplot(data = df)
```
